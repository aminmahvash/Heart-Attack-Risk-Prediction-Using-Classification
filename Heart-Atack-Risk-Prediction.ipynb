# Load Data
# pre-processing and selection

```python
import pandas as pd

df = pd.read_csv('Data/HeartAtackRisk_Data.csv')
print(df.head())
```

```python
df.dtypes
```

```python
import numpy as np
x = np.asarray(df[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh', 'exng', 'caa']])
x
```

```python
y = np.asarray(df['output'])
y
```

# train/test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=34)
print("train shape: ", x_train.shape , y_train.shape)
print("test shape: ", x_test.shape , y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# Modeling
we have four model to train this dataset
and then we choose the better one
1. **KNN**
2. **Decision Tree**
3. **Logistic Regression**
4. **SVM**

## 1. KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(26)
knn = knn.fit(x_train, y_train)
yhat = knn.predict(x_test)
print(yhat)
```

```python
from sklearn import metrics
print("KNN Accuracy: ", metrics.accuracy_score(y_test,yhat))

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, yhat)
print(cm)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

Ks = 40
mean_acc = np.zeros((Ks-1))

for n in range(1, Ks):
    neigh = KNeighborsClassifier(n_neighbors = n).fit(x_train, y_train)
    yhat = neigh.predict(x_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

print( "the best Accuracy:", mean_acc.max(), "is for k = ", mean_acc.argmax()+1)
```

## 2. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
```

```python
drugTree = drugTree.fit(x_train,y_train)
yhat2 = drugTree.predict(x_test)
```

```python
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, yhat2))
cm1 = confusion_matrix(y_test, yhat2)
print(cm1)
```

## 3. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(C=0.1, solver='liblinear').fit(x_train,y_train)

yhat3 = LR.predict(x_test)

yhat_prob = LR.predict_proba(x_test)
```

```python
print("Logistic Regression Accuracy: " ,metrics.accuracy_score(y_test, yhat3))
cm2 = confusion_matrix(y_test, yhat3)
print(cm2)
```

## 4. SVM

```python
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(x_train, y_train)

yhat4 = clf.predict(x_test)
yhat4[0:5]
```

```python
print("SVM Accuracy: " ,metrics.accuracy_score(y_test, yhat4))
cm4 = confusion_matrix(y_test, yhat4)
print(cm4)
```
