# Load Data
# pre-processing and selection

```python
import pandas as pd

df = pd.read_csv('Data/HeartAtackRisk_Data.csv')
print(df.head())
```

```python
df.dtypes
```

```python
import numpy as np
x = np.asarray(df[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh', 'exng', 'caa']])
x
```

```python
y = np.asarray(df['output'])
y
```

# train/test

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(x).transform(x.astype(float))
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=34)
print("train shape: ", x_train.shape , y_train.shape)
print("test shape: ", x_test.shape , y_test.shape)
```

# Modeling
we have four model to train this dataset
and then we choose the better one
1. **KNN**
2. **Decision Tree**
3. **Logistic Regression**
4. **SVM**

## 1. KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(32)
knn = knn.fit(x_train, y_train)
yhat = knn.predict(x_test)
print(yhat)
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, knn.predict(x_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test,yhat))

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, yhat)
print(cm)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

Ks = 40
mean_acc = np.zeros((Ks-1))

for n in range(1, Ks):
    # تمرین مدل برای هر K
    neigh = KNeighborsClassifier(n_neighbors = n).fit(x_train, y_train)
    yhat = neigh.predict(x_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

print( "بهترین دقت:", mean_acc.max(), "در مقدار K=", mean_acc.argmax()+1)
plt.show()
```
